# Different ways to create dataframe

### using excel

```python
df = pd.read_excel('weather_data.xlsx', 'Sheet1')
df
```

### using tuples and list

```python
weather_data = [
    ('1/1/2023', 32, 6, 'Rain'),
    ('1/2/2023', 35, 6, 'Sunny'),
    ('1/3/2023', 32, 2, 'Snow')
]

df = pd.DataFrame(data=weather_data)

df
```

```python
weather_data = [
    ('1/1/2023', 32, 6, 'Rain'),
    ('1/2/2023', 35, 6, 'Sunny'),
    ('1/3/2023', 32, 2, 'Snow')
]

df = pd.DataFrame(data=weather_data, columns=['day', 'temperature', 'windspeed', 'event'])

df
```

### list of dictionary

```python
weather_data = [
    {'day': '1/1/2023', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2023', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2023', 'temperature': 32, 'windspeed': 6, 'event': 'Snow'}
]

df = pd.DataFrame(data=weather_data)
df
```
